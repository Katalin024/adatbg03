```sql
select v.usernev,
        METTOL,
        f.FELNOTT_SZAM
from Vendeg v join Foglalas f on v.USERNEV=f.UGYFEL_FK
group by rollup (USERNEV, FELNOTT_SZAM, METTOL)
order by USERNEV asc, METTOL
```

<span style="color: rgb(0, 0, 0); font-family: &quot;Source Sans 3&quot;, Inter, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">Listázzuk azoknak a vendégeknek a felhasználónevét, akik olyan szálláshelyen foglaltak, ahol a legtöbb csillag van megadva!</span>  

- A megoldást beágyazott lekérdezéssel oldja meg, más esetben legfeljebb 50% pont jár. 
- Az eredményt felhasználónév szerint csökkenő sorrendben jelenítsük meg!

```sql
select distinct usernev
from Vendeg v join Foglalas f on v.usernev=f.UGYFEL_FK
        join szoba sz on f.SZOBA_FK=sz.SZOBA_ID
        join szallashely szh on sz.SZALLAS_FK=szh.SZALLAS_ID
where szh.CSILLAGOK_SZAMA = (
select max(CSILLAGOK_SZAMA)
from szallashely)
order by USERNEV desc
```

<span style="color: rgb(0, 0, 0); font-family: &quot;Source Sans 3&quot;, Inter, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">Listázzuk azoknak a vendégeknek a felhasználónevét, akik foglaltak 2 fős, de nem foglaltak 4 fős szobát!</span>  

- A megoldást halmazműveletek (UNION, INTERSECT, EXCEPT) segítségével oldja meg, más esetben csak 50% pont jár. 
- Rendezzük név szerint az eredményt!

```sql
select distinct usernev
from Vendeg v join Foglalas f on v.usernev=f.UGYFEL_FK
        join szoba sz on f.SZOBA_FK=sz.SZOBA_ID
where sz.FEROHELY = 2 

EXCEPT

select distinct usernev
from Vendeg v join Foglalas f on v.usernev=f.UGYFEL_FK
        join szoba sz on f.SZOBA_FK=sz.SZOBA_ID
where sz.FEROHELY = 4
order by USERNEV asc
```

<span style="box-sizing: border-box; color: rgb(0, 0, 0); font-family: &quot;Source Sans 3&quot;, Inter, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255); border-radius: 0px !important;"><span data-start="214" data-end="242" style="box-sizing: border-box; border-radius: 0px !important;">Listázzuk, hogy a 2016-ban</span>&nbsp;szállást foglaló&nbsp;<span data-start="260" data-end="278" style="box-sizing: border-box; border-radius: 0px !important;">aktív vendégek</span></span> <span style="color: rgb(0, 0, 0); font-family: &quot;Source Sans 3&quot;, Inter, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">&nbsp;esetében típusonként és klíma szerint külön-külön mennyi foglalás történt!</span>  

- Csak egy lekérdezést használjunk

```sql
select usernev,
        sz.FEROHELY,
        sz.KLIMAS,
        sz.POTAGY,
        count(USERnev) as 'típusonkénti foglalásszám'
from Vendeg v join Foglalas f on v.usernev=f.UGYFEL_FK
        join szoba sz on f.SZOBA_FK=sz.SZOBA_ID
where year(METTOL)=2016
group by rollup(usernev, FEROHELY, KLIMAS, POTAGY)
```

<span style="color: rgb(0, 0, 0); font-family: &quot;Source Sans 3&quot;, Inter, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">Számoljuk meg szálláshelyenként, hogy hány különböző szobatípus (TIPUS mező) van!</span>  

- Rendezzük az eredményt csökkenő sorrendben a típusszám szerint!

```sql
select SZALLAS_ID,
        count(distinct TIPUS) as 'tipusszám'
from Szoba sz join szallashely szh on sz.SZALLAS_FK=szh.SZALLAS_ID
group by SZALLAS_ID
order by count(distinct TIPUS) desc
```
